## **Transformación de datos**

### **Conversión de cantidad de reviews**

Se transforma el número de reviews en una variable categórica que indique si la propiedad recibió reseñas de usuarios (Yes) o si este valor es cero (No)

```
# Agrego una nueva columna para dividir entre los que tienen reviews y los que no
datos2020 = datos2020.assign(reviews_if = np.where(datos2020['number_of_reviews'] > 0, 'Yes', 'No'))
datos2021 = datos2021.assign(reviews_if = np.where(datos2021['number_of_reviews'] > 0, 'Yes', 'No'))


print(f"Valores 2020:\n{datos2020['reviews_if'].value_counts()}")
print(f"\nValores 2021:\n{datos2021['reviews_if'].value_counts()}")
```

### **Cálculo del valor total de estadía**

Debido a que los precios son unitarios, por noche, y a que hay una variable relacionada a las condiciones del alquiler (minimum_nights, estadía mínima a contratar), decidimos agregar en el análisis el costo total de la estadía, definido de esta manera:

valor_estadia = minimum_nights * price

```
# Agregado del valor de la estadía a los dos dataframes de trabajo
datos2020 = datos2020.assign(valor_estadia = datos2020['minimum_nights']*datos2020['price'])
datos2021 = datos2021.assign(valor_estadia = datos2021['minimum_nights']*datos2021['price'])
```

**Resultado**

Estructura de los dos dataframes de trabajo con las nuevas columnas reviews_if y valor_estadia

```
archivo['Tabla_datos_2020'] = datos2020
archivo['Tabla_datos_2021'] = datos2021
```

## 3) **Precios según Estadía**
Analisis entre 2020-2021

Se realiza el estudio del promedio del valor de la estadía en relación a los grupos de barrios.

```
promedio2020 = datos2020.groupby('neighbourhood_group')[['valor_estadia','minimum_nights']].mean()
print(promedio2020)
promedio2021 = datos2021.groupby('neighbourhood_group')[['valor_estadia','minimum_nights']].mean()
print(promedio2021)
```

Comparación del valor de la estadía entre los dos períodos

```
# Combinación de los dos dataframes
merge = promedio2020.merge(promedio2021, on='neighbourhood_group',how='inner')

# Cálculo de la variación porcentual entre períodos para el valor de la estadía y la estadía mínima de noches
merge['porcentaje_estadia'] = round((((merge['valor_estadia_y']-merge['valor_estadia_x'])/merge['valor_estadia_y'])*100),2)
merge['porcentaje_min_nights'] = round((((merge['minimum_nights_y']-merge['minimum_nights_x'])/merge['minimum_nights_y'])*100),2)

# Modificación del nombre de las columnas
merge.rename(columns={'valor_estadia_x' : 'valor_estadia_2020', 'valor_estadia_y' : 'valor_estadia_2021' 
                      ,'minimum_nights_x' : 'minimum_nights_2020','minimum_nights_y' : 'minimum_nights_2021'  }, inplace=True)

merge.sort_values('valor_estadia_2021', inplace=True)
merge
```

```
print(f"Promedio de aumento de valor estadía: {round(merge.porcentaje_estadia.mean())}%")
print(f"Promedio de aumento de días de estadía: {round(merge.porcentaje_min_nights.mean())}%")
```

```
grafico = pd.DataFrame({'2020': merge['valor_estadia_2020'],'2021': merge['valor_estadia_2021']})
grafico.plot.bar(title = 'Promedio de valor de la estadía\n')
```

```
grafico = pd.DataFrame({'2020': merge['minimum_nights_2020'],'2021': merge['minimum_nights_2021']})
grafico.plot.bar(title = 'Promedio de días mínimos de estadía\n')
```

**Almacenamiento**

Se guarda la tabla de comparación de valores

```
archivo['Comparacion_valores_estadia'] = merge
```

## 4) **Precios según Reviews**
Analisis entre 2020-2021

```
#contador
datos2020['anio'] = 2020
datos2021['anio'] = 2021
datos2020rev = datos2020.groupby(['anio','reviews_if']).agg({'reviews_if':'count','price':'mean'})
datos2020round = datos2020rev.round(2)

datos2021rev = datos2021.groupby(['anio','reviews_if']).agg({'reviews_if':'count','price':'mean'})
datos2021round = datos2021rev.round(2)

datosreviews = pd.concat([datos2020round,datos2021round], axis=0)
datosreviews
```

```
datos2020room = datos2020.groupby(['anio','reviews_if','room_type']).agg({'reviews_if':'count','price':'mean'})
datos2020roomround = datos2020room.round(2)

datos2021room = datos2021.groupby(['anio','reviews_if','room_type']).agg({'reviews_if':'count','price':'mean'})
datos2021roomround = datos2021room.round(2)

datosroom = pd.concat([datos2020roomround,datos2021roomround], axis=0)
print('Cantidad de Registros que tienen reviews por año según tipo de habitación')
datosroom
```

```
# Reviews con más info
datos2020info = datos2020.groupby(['anio','reviews_if','neighbourhood_group']).agg({'reviews_if':'count','price':'mean'})
datos2020roundinfo = datos2020info.round(2)

datos2021info = datos2021.groupby(['anio','reviews_if','neighbourhood_group']).agg({'reviews_if':'count','price':'mean'})
datos2021roundinfo = datos2021info.round(2)

datosreviewsinfo = pd.concat([datos2020roundinfo,datos2021roundinfo], axis=0)
print('Cantidad de Registros que tienen reviews por año por barrios con su precio medio')
datosreviewsinfo
```

**Almacenamiento**

Se guardan las tres tablas de resultados

```
archivo['Reviews_20_21'] = datosreviews
archivo['Reviews_tipo_habitacion'] = datosroom
archivo['Reviews_barrio_20_21'] = datosreviewsinfo
```
